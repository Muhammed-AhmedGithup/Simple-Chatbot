```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model=AutoModelForSeq2SeqLM.from_pretrained("facebook/blenderbot-400M-distill")
ticonizer=AutoTokenizer.from_pretrained("facebook/blenderbot-400M-distill")
```

```python
def chat_bot():
  while(True):
    text=input("You: ")
    if text in ['bye','good bye' , 'see you letter']:
      break
    inputs=ticonizer(text,return_tensors='pt')
    reply_ids=model.generate(**inputs)
    out=ticonizer.decode(reply_ids,skip_special_tokens=True)
    print("Bot: ",out[0])
```

```python
chat_bot()
```
